```python
# write code here

def update_phone_number(number):
    phone_no = 0
    number = str(number)
    
    if len(number) == 16:
        update_phone_number = "0"+ number[4:]
        phone_no += 1  
        return update_phone_number
    else:
        update_phone_number = "0"+ number
        phone_no += 1  
        return update_phone_number

recweek["phonenumber"] = recweek["phonenumber"].apply(update_phone_number)

recweek
```

```python
# write code here
carrier_df = pd.DataFrame(final_recweek[["phonenumber","gender"]])

def number_to_carrier(number):
    phone_no = 0
    number = str(number)
    
    if number[0:4] == "0918":
        number = "Smart"
        phone_no += 1
        return number
    else:
        number = "Globe"
        phone_no += 1
        return number
        
carrier_df["phonenumber"] = carrier_df["phonenumber"].apply(number_to_carrier)

carrier_df
```
